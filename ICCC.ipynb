```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from transformers import pipeline
import plotly.express as px

# SLoad and combine CSV files into a single DataFrame
file_links = [
    "https://drive.google.com/uc?id=1IXPRcbf77spmAYw1M-XlAXi_bgdFEIG2",
    "https://drive.google.com/uc?id=1Oip8cVOXO9ytK152NUKQ24PYHKndWU97",
    "https://drive.google.com/file/d/1IXPRcbf77spmAYw1M-XlAXi_bgdFEIG2/view?usp=sharing",
"https://drive.google.com/file/d/1Oip8cVOXO9ytK152NUKQ24PYHKndWU97/view?usp=sharing",
"https://drive.google.com/file/d/1x25szx_F5Mf5EMdzw7k-Z3anqh90WJKU/view?usp=sharing",
"https://drive.google.com/file/d/1oURqPOFYiGbmhJuSDeNvXfeEUNemZRuS/view?usp=sharing",
"https://drive.google.com/file/d/1H-SRzhVKZJCe3LuwnHIcAJ1Kd8hodCox/view?usp=sharing",
"https://drive.google.com/file/d/1ODqTo09eMeHljvOAGlVPipvaYnagCAgA/view?usp=sharing",
"https://drive.google.com/file/d/16SNGfePjGbrlHZpjejE7mwUahuCrcYxE/view?usp=sharing",
"https://drive.google.com/file/d/13sAgGnb-g5lh-EUkfE6kC8tj_UiSnb8G/view?usp=sharing",
"https://drive.google.com/file/d/1y_MUHOC-3w_FOe4w2a6Hr5pV8ttTHDl7/view?usp=sharing",
"https://drive.google.com/file/d/17IcYYGyIDeZ79BeHqM_aAoW_ZgviqN8T/view?usp=sharing",
"https://drive.google.com/file/d/1TPcD2d_KKXIWd138U0XSp8UgjH-i1qsp/view?usp=sharing",
"https://drive.google.com/file/d/15LTTD9loVAWBfs-RbR0rrEcwYusS4_T0/view?usp=sharing",
"https://drive.google.com/file/d/1eKtZl0uoAaiiG8R2bSLsHOSRxtAJYaaD/view?usp=sharing"

]
dataframes = []
```

```python
for link in file_links:
    try:
        df = pd.read_csv(link, delimiter=',', engine='python')
        dataframes.append(df)
        print(f"Loaded file: {link}")
    except Exception as e:
        print(f"Error loading file: {link} - {e}")
```

```python

if dataframes:
    crick_df = pd.concat(dataframes, ignore_index=True)
    print("Successfully combined all files into a single DataFrame.")
else:
    print("No valid files loaded.")


crick_df = pd.concat(dataframes, ignore_index=True)



```

```python

#   cleaning part

print(crick_df.info())
crick_df.drop_duplicates(inplace=True)
crick_df.dropna(subset=['date', 'team_1', 'team_2'], inplace=True)

#  Handle outliers and missing values
# Add match_status column

crick_df['match_status'] = crick_df['result'].apply(
    lambda x: 'abandoned' if pd.isna(x) or 'abandoned' in str(x).lower() else 'played'
)

# Add winning_team column
crick_df['winning_team'] = crick_df['result'].apply(
    lambda x: None if 'abandoned' in str(x).lower() else (
        str(x).split(' ')[0] if 'won' in str(x) else None
    )
)
```

```python

def split_players(stats_list, role):
    if pd.isna(stats_list):
        return [None, None]
    stats = eval(stats_list)
    return [f"{player.split(' - ')[0]} ({role})" for player in stats]

crick_df[['best_batter_1', 'best_batter_2']] = crick_df['best_batters'].apply(
    lambda x: split_players(x, 'Batter')).tolist()
crick_df[['best_bowler_1', 'best_bowler_2']] = crick_df['best_bowlers'].apply(
    lambda x: split_players(x, 'Bowler')).tolist()


crick_df.drop(['commentary_line', 'Unnamed: 0.1', 'Unnamed: 0'], axis=1, inplace=True, errors='ignore')

crick_df.to_csv("cleaned_world_cup_data.csv", index=False)


```

```python

# Load commentary_2023.csv
commentary_df = pd.read_csv("https://drive.google.com/uc?id=1g588SKp85SPdthT_SB3qDMhral2E1VBG")

# Load a Hugging Face sentiment analysis model
sentiment_analyzer = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
```

```python

# Perform sentiment analysis
commentary_df['sentiment'] = commentary_df['commentary'].apply(
    lambda x: sentiment_analyzer(x)[0]['label']
)

# Visualize sentiment spread
fig = px.pie(commentary_df, names='sentiment', title='Sentiment Spread for 2023 World Cup Final')
fig.show()
```

```python

commentary_df.to_csv("commentary_with_sentiments.csv", index=False)
```

```python
import dash
from dash import dcc, html
import plotly.express as px
import pandas as pd

# Sample data for demonstration
data = {
    "world_cup_year": [1975, 1979, 1983, 1987, 1992],
    "winning_team": ["West Indies", "West Indies", "India", "Australia", "Pakistan"],
    "match_status": ["played", "played", "abandoned", "played", "played"]
}


crick_df = pd.DataFrame(data)




```

```python
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("ICC Men's Cricket World Cup Dashboard", style={'textAlign': 'center'}),
    dcc.Graph(
        id='winning-teams-bar',
        figure=px.bar(crick_df, x='world_cup_year', y='winning_team',
                      title='Winning Teams Over the Years',
                      labels={'world_cup_year': 'Year', 'winning_team': 'Winning Team'})
    ),
    dcc.Graph(
        id='match-status-pie',
        figure=px.pie(crick_df, names='match_status', title='Match Status Distribution')
    ),
])
```

```python

if __name__ == '__main__':
    app.run_server(debug=True)
```
